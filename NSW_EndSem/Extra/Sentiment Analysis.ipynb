```python
import igraph as ig
import networkx as nx
from textblob import TextBlob
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

pd.options.display.max_rows = 999
```

```python
twitter_data = pd.read_csv('/home/vivek/Desktop/NSW/Project/Data/#GameofThrones_tweets.csv')
twitter_data.head()
```

```python
comment_words = ' ' #We will be appending the words to this var
stopwords = set(STOPWORDS) #Finds all stop words in the set of tweets.
for val in twitter_data.text:
   val = str(val) #convert all tweet content into strings
   tokens = val.split() #Split all strings into individual components
   for i in range(len(tokens)):
      tokens[i] = tokens[i].lower() #Converts all the individual strings to lower case.
for words in tokens:
   comment_words = comment_words + words + ' '
```

```python
#comment_words=' '
#stopwords=set(STOPWORDS)
polarity = []
subjectivity = []
# print('  Polarity  |   Subjectivity  |   Tweet')
for val in twitter_data.index:
    sentiments = TextBlob(twitter_data['text'][val])
    polarity.append(sentiments.polarity)
    subjectivity.append(sentiments.subjectivity)
#     print('---------------')
#     print(sentiments.polarity,end='')
#     print(' |  ',end='')
#     print(sentiments.subjectivity, end='')
#     print(' | ',end='')
#     print(twitter_data['text'][val])
```

```python
final = pd.DataFrame(twitter_data['text'])
```

```python
subVal = []
polarityVal = []
for i in polarity:
    if i>0:
        polarityVal.append('Positive')
    elif i==0:
        polarityVal.append('Neutral')
    else:
        polarityVal.append('Negative')

for i in subjectivity:
    if i > 0.5:
        subVal.append('Subjective')
    else:
        subVal.append('Objective')
```

```python
final['Polarity'] = polarityVal
final['Subjectivity'] = subVal
```
