```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from datetime import datetime
import math

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
tweetsCSV = pd.read_csv('/home/tripti/Desktop/Vivek/NSW/Project/GOT/#GameofThrones_tweets.csv')
cols = tweetsCSV.columns
data = tweetsCSV[['user_id',
                  'created_at', 
                  'screen_name',
                  'status_id',
                  'retweet_count', 
                  'retweet_created_at', 
                  'retweet_user_id',
                  'retweet_screen_name',
                  'retweet_status_id']]
data = data.dropna()
data.sort_values('user_id')
```

```python
data['created_at'] = [ d.replace('-', '/') for d in data['created_at']]
data['retweet_created_at'] = [ str(d).replace('-', '/') for d in data['retweet_created_at']]
data['created_at'] = [ datetime.strptime(d, '%Y/%m/%d %H:%M:%S') for d in data['created_at']]
data['retweet_created_at'] = [ datetime.strptime(d, '%Y/%m/%d %H:%M:%S') for d in data['retweet_created_at']]
data['Date'] = [d.date() for d in data['created_at']]
data['Time'] = [d.time() for d in data['created_at']]
data['Date_retweet'] = [d.date() for d in data['retweet_created_at']]
data['Time_retweet'] = [d.time() for d in data['retweet_created_at']]
data.drop(['created_at', 'retweet_created_at'], axis = 1)
data.head()
```
