```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/home/tripti/Desktop/Vivek/NSW/Project/GOT/GOTRetweetEdgeList.csv')
G = nx.DiGraph()
for i in data.index:
    u = str(data['source'][i])
    v = str(data['target'][i])
    G.add_edge(v, u)
    
def plotGraph(G, values):
    layout = nx.spring_layout(G, k = 0.4)
    figsize = (25, 25)
    fig = plt.figure(figsize=figsize)
                         
    plt.axis("off")
    nx.draw_networkx(G, pos = layout, 
                     cmap = plt.get_cmap("jet"),
                     node_color = 'r',
                     node_size = values, 
                     with_labels = True
                    )
```

# Page Rank

```python
pagerank = nx.pagerank(G)
values = list(pagerank.values())
values = [val*1e5 for val in values]
plotGraph(G, values)
```

```python
degreeCentrality = nx.degree_centrality(G)
values = list(degreeCentrality.values())
values = [val*1e4 for val in values]
plotGraph(G, values)
```

```python
eigenvectorCentrality = nx.eigenvector_centrality(G, max_iter=1000, tol=1e-06)
values = list(eigenvectorCentrality.values())
values = [val*1e4 for val in values]
plotGraph(G, values)
```

```python
closenessCentrality = nx.closeness_centrality(G)
values = list(closenessCentrality.values())
values = [val*1e5 for val in values]
plotGraph(G, values)
```

```python
betweennessCentrality = nx.betweenness_centrality(G)
values = list(betweennessCentrality.values())
values = [val*1e7+10 for val in values]
plotGraph(G, values)
```

```python
transitivity = nx.transitivity(G)
```
