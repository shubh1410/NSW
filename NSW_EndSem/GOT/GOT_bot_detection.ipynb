```python
# GOT dataset
import csv
import pandas as pd
from itertools import product
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import csv
data_follower=pd.read_csv('/home/shuham/Desktop/Projects/NSW/NSW_Kashmir/#GameofThrones_follower_followee_network.csv')
data_tweet= pd.read_csv('/home/shuham/Desktop/Projects/NSW/NSW_Kashmir/#GameofThrones_tweets.csv')
```

```python
filtered_data = data_tweet[['user_id','screen_name','lang','followers_count','friends_count','profile_url','description','location']]
#filtered_data=filtered_data.fillna(0)
filtered_data['ratio']=filtered_data['followers_count']/filtered_data['friends_count']
filtered_data
```

```python
for i in range(200):
    if filtered_data['ratio'][i] >=0.45 and filtered_data['ratio'][i]<=0.55:
        filtered_data['ratio'][i]=1
    else:
        filtered_data['ratio'][i]=0

filtered_data['user_id'] = np.where(filtered_data['user_id'].isnull(), 0, 1)
filtered_data['screen_name'] = np.where(filtered_data['screen_name'].isnull(), 0, 1)
filtered_data['lang'] = np.where(filtered_data['lang']!="en", 0, 1)
filtered_data['followers_count'] = np.where(filtered_data['followers_count']>1000,0,filtered_data['followers_count'])
filtered_data['followers_count'] = np.where(filtered_data['followers_count']<30,0,1)

filtered_data['friends_count'] = np.where(filtered_data['friends_count']>1200, 0, 1)
filtered_data['profile_url'] = np.where(filtered_data['profile_url'].isnull(), 0, 1)
filtered_data['description'] = np.where(filtered_data['description'].isnull(), 0, 1)
filtered_data['location'] = np.where(filtered_data['location'].isnull(), 0, 1)
```

```python
filtered_data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 2, init='k-means++',max_iter=1000)
kmeans.fit(X=filtered_data)
pred = kmeans.predict(filtered_data)
kmeans.inertia_
```

```python
frame = pd.DataFrame(filtered_data) 
frame['cluster'] = pred 
frame['cluster'].value_counts()
```

```python
frame.to_csv('/home/shuham/Desktop/data1_GOT.csv')
```
