```python
import networkx as nx
import matplotlib.pyplot as plt
import operator
from networkx.algorithms.approximation import max_clique
```

```python
#http://konect.uni-koblenz.de/networks/dolphins
g=nx.read_edgelist('../dolphins/out.dolphins',create_using=nx.Graph(),nodetype=int)
nx.draw(g,node_size=200,with_labels=True)
```

a. Weiner Index

```python
#http://konect.uni-koblenz.de/networks/dolphins​
g=nx.read_edgelist('../dolphins/out.dolphins',create_using=nx.Graph(),nodetype=int)
print(nx.wiener_index(g))
g.remove_node(35)
print(nx.wiener_index(g))
g.remove_node(2)
print(nx.wiener_index(g))
```

b. Max Clique

```python
#http://konect.uni-koblenz.de/networks/dolphins
g=nx.read_edgelist('../dolphins/out.dolphins',create_using=nx.Graph(),nodetype=int)
s = max_clique(g)
list=[]
for i in range(1,63):
    if(i!=19 or i!=25 or i!=30 or i!=46 or i!=52):
        list.append(i)
nx.draw_circular(g,nodelist=list,node_color='b',with_labels=True)
nx.draw_circular(g,nodelist=s,node_color='y',with_labels=True)
```

c. N-Plex

```python
#http://konect.uni-koblenz.de/networks/dolphins
g=nx.read_edgelist('../dolphins/out.dolphins',create_using=nx.Graph(),nodetype=int)
remove_nodes = [node for node,degree in g.degree() if degree < len(g) - 60]
g.remove_nodes_from(remove_nodes)
nx.draw_circular(g,nodelist=g.nodes(),node_color='b',with_labels = True, node_size = 200)
```

d. Core

```python
#http://konect.uni-koblenz.de/networks/dolphins
g=nx.read_edgelist('../dolphins/out.dolphins',create_using=nx.Graph(),nodetype=int)
print(nx.core_number(g))
graph=(nx.k_core(g))
nx.draw_circular(graph,with_labels=True)
```
